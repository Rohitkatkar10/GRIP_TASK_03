step2: Removal of redundunt data.

```python
given_data.duplicated().sum()
```

```python
given_data.drop_duplicates(keep = 'first', inplace = True)
given_data.shape
```

```python
given_data.describe()
```

```python
corelation = given_data.corr()
corelation
```

```python
#as country is same i will drop it.
df = given_data.drop(['Country', 'Postal Code'], axis = 1)
df.head()
```

step3: Relatioship Analysis.

```python
corelation_df = df.corr()
corelation_df
```

```python
sns.heatmap(corelation_df, annot = True)
```

```python
df['State'].value_counts()
```

```python
#plotting graph for states
plt.figure(figsize=(15,15))
sns.countplot(x = df['State'])
plt.xticks(rotation = 90)
plt.title('state count')
plt.show
```

```python
df['Sub-Category'].value_counts()
plt.figure(figsize = (20, 20))
sns.countplot(x = df['Sub-Category'])
plt.xticks(rotation = 90)
plt.title('Sub-category count')
plt.show
```

```python
plt.figure(figsize = (20, 20))
sns.barplot(x = df['Sub-Category'], y = df['Profit'])
plt.title('profit vs sub-category')
plt.show
```

```python
plt.figure(figsize = (20,20))
sns.pairplot(df, hue = 'Sub-Category')
plt.show
```

```python
plt.figure(figsize = (20,20))
sns.pairplot(df, hue = 'Region')
plt.show
```

```python
plt.figure(figsize = (20,20))
sns.pairplot(df, hue = 'Segment')
plt.show
```

```python
df.plot(x = 'Sales', y = 'Profit', style = 'x')
plt.title('profit vs sales')
plt.xlabel('sales')
plt.ylabel('profit')
plt.legend()
plt.show()
```

```python
df.plot(x = 'Region', y = 'Profit')  #width = '5')
plt.title('Profit vs Region')
plt.xlabel('Region')
plt.ylabel('profit')
plt.legend()
plt.show()
```
